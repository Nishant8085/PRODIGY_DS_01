```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("population.csv")
```

```python
data.head(267)
```

```python
#List of contries for the graph.
country = ['United States', 'Canada', 'China', 'Japan', 'Germany', 'India']

#Filtering the data frame
filtered_data = data[data['Country Name'].isin(country)]

#Creating the bar chart
plt.bar (filtered_data['Country Name'], filtered_data['2021'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2021')
plt.title('Population of selected countries in 2021')
plt.xticks(rotation=45, ha='right') #Rotating the x-axis labels for readability
plt.show()
```
